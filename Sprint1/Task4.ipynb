```polyglot-notebook
public class Product
{
    private string _name;
    private decimal _price;
    private int quanitty;
    protected string category;
    internal string description;

    public string Name 
    {
        get{return _name;}
        set{_name = value;}
    }
    public decimal Price
    {
        get{return _price;}
        set {
            if (value<0)
                throw new ArgumentOutOfRangeException("Цена не может быть отрицательной");
            else {_price=value;}
        }   
    }
    public int Quanitty 
    {
        get {return quanitty;}
        set {
            if (value<0)
                throw new ArgumentOutOfRangeException("Кол-во не может быть отрицательной");
            else quanitty=value;
        }
    }
    public string Category 
    {
        get{return category;}
        set{category = value;}
    }
    public string Description 
    {
        get{return description;}
        set{description=value;}
    }

    public decimal GetTotalPrice()
    {
        decimal total = _price*quanitty;
        return total;
    }
    public void UpdateQuanitty(int amount)
    {
        quanitty += amount;
    }
    public string GetProductInfo()
    {
        return $"Название товара: {Name} ;  Цена: {Price} ; Кол-во: {Quanitty}; Категория: {Category}; Описание: {Description}";
    }
}

public class Seller 
{
    private string name ;
    private string employeeld;
    protected decimal salary;
    internal string contactinfo;
    private List<Product> products = new List<Product>();

    public string NameSel 
    {
        get{return name;}
        set{name = value;}
    }
    public string Eploy 
    {
        get{return employeeld;}
        set{employeeld = value;}
    }
    public decimal Salary 
    {
        get{return salary;}
        set{
            if (value >= 0)
                salary = value;
            else throw new ArgumentOutOfRangeException("Заработная плата не может быть отрицательной");    
            }
    }
    public string ContactInfo 
    {
        get{return contactinfo;}
        set{contactinfo = value;}
    }

    public void AddProduct(Product product) 
    {
        products.Add(product);
    }
    public void SellProduct(Product product, int quanitty)
    {
        product.UpdateQuanitty(-quanitty);
    }
    public string getSellerInfo()
    {
        return $"Продавец: {NameSel} ; Таб.№: {Eploy} ; Зар. плата: {Salary} ; Инфо.: {ContactInfo}";
    }
    public List<Product> GetProducts()
    {
        return products;
    }
}

public class Store
{
    private string storeName;
    private string location;
    public string storeHours;
    public string StoreName{get{return storeName;}set{storeName=value;}}
    public string Location{get{return location;}set{location=value;}}
    public string StoreHours {get{return storeHours;}set{storeHours=value;}}
    
    private List<Seller> sellers = new List<Seller>();

    public void AddSeller(Seller seller)
    {
        sellers.Add(seller);
    }
    public void ListProduct()
    {
        foreach (var seller in sellers)
        {
            foreach(var product in seller.GetProducts())
            {
                Console.WriteLine(product.GetProductInfo());
            }
        }
    }
    public string GetStoreInfo()
    {
        return $"Название: {StoreName} ; Расположение: {Location} ; Время работы: {StoreHours}";
    }
}

Product product1 = new Product();
product1.Name = "Ананас";
product1.Price = 450.0m;
product1.Quanitty = 12;
product1.Category = "Фрукты";
product1.Description = "Вкусный";
Console.WriteLine(product1.GetProductInfo());
Console.WriteLine($"Общая стоимость всего товара : {product1.GetTotalPrice()}");
product1.UpdateQuanitty(12);
Console.WriteLine(product1.GetProductInfo());

Product product2 = new Product();
product2.Name = "Ананас розовый";
product2.Price = 450.0m;
product2.Quanitty = 8;
product2.Category = "Фрукты и овощи";
product2.Description = "Вкусный клянусь";
Console.WriteLine(product2.GetProductInfo());
Console.WriteLine($"Общая стоимость всего товара : {product2.GetTotalPrice()}");
product2.UpdateQuanitty(2);
Console.WriteLine(product2.GetProductInfo());

// Пример вывода для Seller
Seller seller1 = new Seller();
seller1.NameSel = "Иванов";
seller1.Eploy = "123";
seller1.Salary = 30000;
seller1.ContactInfo = "ivanov@mail.ru";
seller1.AddProduct(product1);
seller1.AddProduct(product2);
Console.WriteLine(seller1.getSellerInfo());
foreach(var product in seller1.GetProducts())
{
    Console.WriteLine(product.GetProductInfo());
}

// Пример вывода для Store
Store store1 = new Store();
store1.StoreName = "Магазин №1";
store1.Location = "ул. Ленина, 1";
store1.StoreHours = "9:00-21:00";
store1.AddSeller(seller1);
Console.WriteLine(store1.GetStoreInfo());
store1.ListProduct();
```
