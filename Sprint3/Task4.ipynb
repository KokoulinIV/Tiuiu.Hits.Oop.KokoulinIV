```polyglot-notebook
using System;
using System.Collections.Generic;

public interface Iinstallable
{
    void Install();
}

public abstract class Component
{
    public string Name { get; set; }
    public decimal Price { get; set; }
    public string Type { get; set; }

    public Component(string name, decimal price, string type)
    {
        Name = name;
        Price = price;
        Type = type;
    }

    public override string ToString()
    {
        return $"Название: {Name}, Цена: {Price}, Тип: {Type}";
    }

    public void Comment(string a)
    {
        Console.WriteLine($"Комментарий: {a}");
    }
    public void Comment(int a)
    {
        Console.WriteLine($"Комментарий: {a.ToString()}");
    }

    public abstract void GetInfo();

   
    public static bool operator ==(Component left, Component right)
    {
        if (ReferenceEquals(left, null))
            return ReferenceEquals(right, null);

        return left.Equals(right);
    }

    
    public static bool operator !=(Component left, Component right)
    {
        return !(left == right);
    }

    
    public override bool Equals(object obj)
    {
        if (obj is Component other)
        {
            return Price == other.Price && 
                   Type == other.Type;
        }
        return false;
    }

    
    public override int GetHashCode()
    {
        return HashCode.Combine(Price, Type);
    }

}

public class CPU : Component, Iinstallable
{
    public int CoreCount { get; set; }
    public float ClockSpeed { get; set; }
    public string TypeInstall { get; set; }

    public CPU(string name, decimal price, string type, int coreCount, float clockSpeed, string typeInstall)
        : base(name, price, type)
    {
        CoreCount = coreCount;
        ClockSpeed = clockSpeed;
        TypeInstall = typeInstall;
    }

    public override void GetInfo()
    {
        Console.WriteLine($"CPU Информация: {ToString()}, Ядер: {CoreCount}, Тактовая частота: {ClockSpeed}GHz");
    }

    public void Install()
    {
        Console.WriteLine($"Установка CPU {TypeInstall}");
    }

    public static bool operator ==(CPU left, CPU right)
    {
        if (ReferenceEquals(left, null))
            return ReferenceEquals(right, null);

        return left.Equals(right);
    }

   
    public static bool operator != (CPU left, CPU right)
    {
        return !(left == right);
    }

    
    public override bool Equals(object obj)
    {
        if (!base.Equals(obj)) return false;
        if (obj is CPU other)
        {
            return CoreCount == other.CoreCount && 
                   ClockSpeed == other.ClockSpeed;
        }
        return false;
    }

    
    public override int GetHashCode()
    {
        return HashCode.Combine(base.GetHashCode(), CoreCount, ClockSpeed);
    }
}
public class Ram : Component, Iinstallable
{
        public int Capacity { get; set; }
        public float Speed { get; set; }
        public string TypeInstall { get; set; }

        public Ram(string name, decimal price, string type, int capacity, float speed, string typeInstall)
            : base(name, price, type)
        {
            Capacity = capacity;
            Speed = speed;
            TypeInstall = typeInstall;
        }

        public override void GetInfo()
        {
            Console.WriteLine($"RAM Информация: {ToString()}, Объём: {Capacity}GB, Скорость: {Speed}MHz");
        }

        public void Install()
        {
            Console.WriteLine($"Установка RAM: {TypeInstall}");
        }

        public static bool operator ==(Ram left, Ram right)
        {
            if (ReferenceEquals(left, null))
                return ReferenceEquals(right, null);

            return left.Equals(right);
        }

        public static bool operator !=(Ram left, Ram right)
        {
            return !(left == right);
        }

        public override bool Equals(object obj)
        {
            if (!base.Equals(obj)) return false;
            if (obj is Ram other)
            {
                return Capacity == other.Capacity &&
                       Speed == other.Speed;
            }
            return false;
        }

        public override int GetHashCode()
        {
            return HashCode.Combine(base.GetHashCode(), Capacity, Speed);
        }
    
}

public class GPU : Component, Iinstallable
{
    public int MemorySize { get; set; }
    public float ClockSpeed { get; set; }
    public string TypeInstall { get; set; }

    public GPU(string name, decimal price, string type, int memorySize, float clockSpeed, string typeInstall)
        : base(name, price, type)
    {
        MemorySize = memorySize;
        ClockSpeed = clockSpeed;
        TypeInstall = typeInstall;
    }

    public override void GetInfo()
    {
        Console.WriteLine($"GPU Информация: {ToString()}, Объём памяти: {MemorySize}GB, Тактовая частота: {ClockSpeed}GHz");
    }

    public void Install()
    {
        Console.WriteLine($"Установка GPU: {TypeInstall}");
    }

    public static bool operator ==(GPU left, GPU right)
    {
        if (ReferenceEquals(left, null))
            return ReferenceEquals(right, null);

        return left.Equals(right);
    }

    public static bool operator !=(GPU left, GPU right)
    {
        return !(left == right);
    }

    public override bool Equals(object obj)
    {
        if (!base.Equals(obj)) return false;
        if (obj is GPU other)
        {
            return MemorySize == other.MemorySize &&
                   ClockSpeed == other.ClockSpeed;
        }
        return false;
    }

    public override int GetHashCode()
    {
        return HashCode.Combine(base.GetHashCode(), MemorySize, ClockSpeed);
    }
}

public class HDD : Component, Iinstallable
{
    public int Capacity { get; set; }
    public float Speed { get; set; }
    public string TypeInstall { get; set; }

    public HDD(string name, decimal price, string type, int capacity, float speed, string typeInstall)
        : base(name, price, type)
    {
        Capacity = capacity;
        Speed = speed;
        TypeInstall = typeInstall;
    }

    public override void GetInfo()
    {
        Console.WriteLine($"HDD Информация: {ToString()}, Объём: {Capacity}GB, Скорость: {Speed}RPM");
    }

    public void Install()
    {
        Console.WriteLine($"Установка HDD: {TypeInstall}");
    }

    public static bool operator ==(HDD left, HDD right)
    {
        if (ReferenceEquals(left, null))
            return ReferenceEquals(right, null);

        return left.Equals(right);
    }

    public static bool operator !=(HDD left, HDD right)
    {
        return !(left == right);
    }

    public override bool Equals(object obj)
    {
        if (!base.Equals(obj)) return false;
        if (obj is HDD other)
        {
            return Capacity == other.Capacity &&
                   Speed == other.Speed;
        }
        return false;
    }

    public override int GetHashCode()
    {
        return HashCode.Combine(base.GetHashCode(), Capacity, Speed);
    }
}


public class ComputerCollection<T> where T : Component
{
    public List<T> components = new List<T>();

    public void AddComponent(T component)
    {
        components.Add(component);
    }
    public void RemoveComponent(T component)
    {
        components.Remove(component);
    }

    public void GetInfo()
    {
        foreach (var component in components)
        {
            component.GetInfo();
        }
    }

    public decimal GetTotalPrice()
    {
        decimal total = 0;
        foreach (var component in components)
        {
            total += component.Price;
        }
        return total;
    }
}


CPU cpu1 = new CPU("Intel Core i7", 300, "LGA 1151", 8, 3.6f, "Внутренний");
CPU cpu3 = new CPU ("Intel Core i7", 300, "LGA 1151", 8, 3.6f, "Внутренний");
CPU cpu2 = new CPU("AMD Ryzen 7", 350, "AMU", 8, 3.8f, "Внутренний");
Ram ram1 = new Ram("Corsair Vengeance LPX", 100, "DIMM", 16, 3200, "Внутренний");
Ram ram2 = new Ram("G.Skill Ripjaws V", 120, "DIMM", 16, 3600, "Внутренний");
GPU gpu1 = new GPU("NVIDIA GeForce RTX 3080", 700, "PCIe", 10, 1.5f, "Внутренний");
GPU gpu2 = new GPU("AMD Radeon RX 6800", 650, "PCIe", 16, 1.8f, "Внутренний");
HDD hdd1 = new HDD("Seagate Barracuda", 50, "SATA", 2000, 7200, "Внутренний");
HDD hdd2 = new HDD("Western Digital Blue", 60, "SATA", 2000, 5400, "Внутренний");

cpu1.Comment("Это комментарий для CPU 1");
cpu2.Comment(42);
 

Console.WriteLine(cpu1 == cpu2);
Console.WriteLine(cpu1 == cpu3);
Console.WriteLine(ram1 == ram2);
Console.WriteLine(gpu1 == gpu2);
Console.WriteLine(hdd1 == hdd2);

ComputerCollection<CPU> cpuCollection = new ComputerCollection<CPU>();
cpuCollection.AddComponent(cpu1);
cpuCollection.AddComponent(cpu2);
ComputerCollection<Ram> ramCollection = new ComputerCollection<Ram>();
ramCollection.AddComponent(ram1);
ramCollection.AddComponent(ram2);
ComputerCollection<GPU> gpuCollection = new ComputerCollection<GPU>();
gpuCollection.AddComponent(gpu1);
gpuCollection.AddComponent(gpu2);
ComputerCollection<HDD> hddCollection = new ComputerCollection<HDD>();
hddCollection.AddComponent(hdd1);
hddCollection.AddComponent(hdd2);

Console.WriteLine($"Стоимость всех CPU: {cpuCollection.GetTotalPrice()}");
Console.WriteLine($"Стоимость всех RAM: {ramCollection.GetTotalPrice()}");
Console.WriteLine($"Стоимость всех GPU: {gpuCollection.GetTotalPrice()}");
Console.WriteLine($"Стоимость всех HDD: {hddCollection.GetTotalPrice()}");

decimal totalOverall = cpuCollection.GetTotalPrice() +
                       ramCollection.GetTotalPrice() +
                       gpuCollection.GetTotalPrice() +
                       hddCollection.GetTotalPrice();
Console.WriteLine($"Стоимость всех компонентов: {totalOverall}");


cpuCollection.GetInfo();
ramCollection.GetInfo();
gpuCollection.GetInfo();
hddCollection.GetInfo();
```
